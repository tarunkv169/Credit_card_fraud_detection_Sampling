```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Creditcard_data.csv')
```

```python
df
```

```python
x = df.drop(['Class'],axis=1)
y = df['Class']
```

```python
x
```

```python
y
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)

x_bal,y_bal = smote.fit_resample(x,y)
```

```python
Z = 1.96
p = 0.5
E = 0.05
S = 5
C = 10
```

```python
n_simple = int((Z**2 * p * (1-p)) / E**2)
n_strat = int((Z**2 * p * (1-p)) / (E/S)**2)
n_clust = int((Z**2 * p * (1-p)) / (E/C)**2)
n_sys = len(x_bal) // 20
```

```python
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    'RandomForest': RandomForestClassifier(),
    'GradientBoost': GradientBoostingClassifier(),
    'LogReg': LogisticRegression(),
    'SVM': SVC(),
    'KNgb': KNeighborsClassifier()
}
```

```python
simple_sample = x_bal.sample(n=n_simple, random_state=42)
```

```python
strat = pd.qcut(y_bal, q=S, labels=False, duplicates='drop')
strat_sample = x_bal.groupby(strat, group_keys=False).apply(
    lambda x: x.sample(min(len(x), n_strat // S), random_state=42)
)
```

```python
kmeans = KMeans(n_clusters=C, random_state=42)
clusters = kmeans.fit_predict(x_bal)
x_bal['cluster'] = clusters
sample_cluster = x_bal.groupby('cluster', group_keys=False).apply(
    lambda x: x.sample(min(len(x), n_clust // C), random_state=42)
)
```

```python
interval = len(x_bal) // n_sys
indices = np.arange(0, len(x_bal), step=interval)
sample_systematic = x_bal.iloc[indices]
```

```python
sample_multistage = strat_sample.groupby(strat, group_keys=False).apply(
    lambda x: x.sample(min(len(x), n_strat // S // 2), random_state=42)
)
```

```python
results = []
sample_names = ['Simple Random', 'Stratified', 'Cluster', 'Systematic', 'Multistage']
samples = [simple_sample, strat_sample, sample_cluster, sample_systematic, sample_multistage]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
for sample_name, sample in zip(sample_names, samples):
    X_sample = sample.drop(['cluster'], axis=1, errors='ignore')  # Drop the 'cluster' column for non-cluster samples
    y_sample = y_bal.loc[X_sample.index]
    X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.2, random_state=42)

    for model_name, model in models.items():
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)
        accuracy = accuracy_score(y_test, predictions)
        results.append({'Model': model_name, 'Sampling Technique': sample_name, 'Accuracy': accuracy})
```

```python
df_results = pd.DataFrame(results)
```

```python
pivot_table = df_results.pivot(index='Model', columns='Sampling Technique', values='Accuracy')
```

```python
pivot_table.to_csv('sampling.csv')
print(pivot_table)
```
